```python
import pandas as pd
import requests
import xml.etree.ElementTree as ET
from datetime import datetime, timedelta

start_date = datetime.strptime('12.07.2012', '%d.%m.%Y')
end_date = datetime.strptime('12.07.2014', '%d.%m.%Y')

data = []

while start_date <= end_date:
    date_str = start_date.strftime('%d.%m.%Y')
    url = f"http://www.cbr.ru/scripts/XML_daily.asp?date_req={date_str}&VAL_NM_RQ=EUR"

    response = requests.get(url)

    if response.status_code == 200:
        root = ET.fromstring(response.content)

        for valute in root.findall('Valute'):
            if valute.find('CharCode').text == 'EUR':
                date = root.attrib['Date']
                value = valute.find('Value').text

                data.append({'Date': date, 'Value': value})

    start_date += timedelta(days=1)

df = pd.DataFrame(data)
print(df)
```

```python
import pandas as pd
import requests

# Define the base URL for fetching exchange rate data
base_url = "https://www.cbr-xml-daily.ru/daily_json.js"
currency_code = "EUR"

# Function to fetch exchange rate data for EUR for the past 10 years
def fetch_exchange_rates(currency_code):
    data = []

    # Fetch data for the past 10 years
    for year in range(2012, 2022):
        url = f"{base_url}?date_req=01/01/{year}"
        response = requests.get(url)

        if response.status_code == 200:
            exchange_rates = response.json()['Valute']
            if currency_code in exchange_rates:
                rate = exchange_rates[currency_code]['Value']
                date = exchange_rates[currency_code]['PreviousDate']
                data.append({'Date': date, 'Rate': rate})

    df = pd.DataFrame(data)
    df['Date'] = pd.to_datetime(df['Date'])  # Convert 'Date' column to datetime type
    df['Rate'] = pd.to_numeric(df['Rate'])  # Convert 'Rate' column to numeric type

    return df

# Call the function to fetch exchange rate data for EUR for the past 10 years
exchange_rates_df = fetch_exchange_rates(currency_code)
print(exchange_rates_df)
```

```python
import pandas as pd
import requests
from datetime import datetime
from statsmodels.tsa.statespace.sarimax import SARIMAX
from matplotlib import pyplot as plt
import xml.etree.ElementTree as ET
from datetime import datetime, timedelta

url = f"http://iss.moex.com/iss/engines/stock/markets/shares/securities/YNDX/candles.json?from=2023-05-25&till=2024-04-03&interval=24"
response = requests.get(url)
data = response.json()
df2 = pd.DataFrame(data['candles']['data'], columns=data['candles']['columns'])

result_active = df2[['begin', 'close']]

model = SARIMAX(result_active['close'], order=(3, 1, 1), seasonal_order=(1, 1, 1, 12))
model_fit = model.fit()

forecast = model_fit.forecast(steps=31)

start_date = '2024-04-04'
end_date = pd.to_datetime(start_date) + pd.DateOffset(days=30)
date_range = pd.date_range(start=start_date, end=end_date, freq='D')
result_yandex1 = {'data': date_range, 'price': forecast}
result_yandex = pd.DataFrame(result_yandex1)
result_yandex
```

```python
import numpy as np
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


train_data, test_data = train_test_split(result_active, test_size=0.2, shuffle=False)

X = train_data['close']
y = test_data['begin']

model = SARIMAX(X, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
model_fit = model.fit()

predictions = model_fit.predict(start=len(X), end=len(X) + len(test_data) - 1, exog=y, dynamic=False)

mae = mean_absolute_error(test_data['close'], predictions)
mse = mean_squared_error(test_data['close'], predictions)
r2 = r2_score(test_data['close'], predictions)
print(mae)
print(mse)
print(r2)
```
